```python
import sqlite3

path = "/Users/marclamy/Library/Application Support/Cursor/User/workspaceStorage/4f77c31c63f9e5367d39876692093d3a/state.vscdb"
conn = sqlite3.connect(f'file:{path}?mode=ro', uri=True)
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

for table in tables:
    table_name = table[0]
    print(f"\nTable: {table_name}")
    cursor.execute(f"PRAGMA table_info('{table_name}');")
    columns = cursor.fetchall()
    print("Columns:")
    for col in columns:
        print(f"  - {col[1]} ({col[2]})")
    cursor.execute(f"SELECT COUNT(*) FROM '{table_name}';")
    row_count = cursor.fetchone()[0]
    print(f"Number of rows: {row_count}")

    # For ItemTable, print first 10 rows (show only length of value, don't try to decode or hex)
    if table_name == "ItemTable":
        print("\nFirst 10 rows from ItemTable:")
        cursor.execute(f"SELECT key, value FROM '{table_name}' LIMIT 10;")
        rows = cursor.fetchall()
        for idx, (k, v) in enumerate(rows):
            print(f"\n  Row {idx+1}:")
            print(f"    Key: {repr(k)}")
            # Value handling: show type & length if bytes, else indicate its type & repr
            if v is None:
                print("    Value: <NULL>")
            elif isinstance(v, bytes):
                print(f"    Value: <BLOB, length={len(v)}>")
            else:
                print(f"    Value: <{type(v).__name__}: {repr(v)}>")
conn.close()
```
